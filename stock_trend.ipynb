```python
# 讀取紀錄各文章所有股票名的檔案
import pandas as pd
import numpy as np

df = pd.read_csv('stock_name.csv')
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

# 計算詞頻
# 不記錄出現次數過少的股票
text_cv = CountVectorizer(min_df=0.005)

stocks = text_cv.fit_transform(df['Stock'])
```

```python
tf_df = pd.DataFrame(stocks.toarray(), columns=text_cv.get_feature_names())
tf_df
```

```python
tf_df['Date'] = df['Date']
tf_df['Website'] = df['Website']
```

```python
tf_df['Date'] = pd.to_datetime(tf_df['Date'], errors = 'coerce')
```

```python
tf_df.to_csv('stock_trend_268cols.csv', index=False)
```

```python
tf = tf_df.drop(columns=['Date','Website'])
tf
```

```python
#counts = pd.DataFrame({'Name':tf_df.sum().index,'Counts':tf_df.sum()})
counts = pd.DataFrame({'Counts':tf.sum()})
counts
```

```python
#pd.set_option('display.max_rows', 100)
counts.sort_values(by=['Counts'],ascending=False).head(50)
```

```python
df202007 = tf_df.loc[tf_df['Date']>=pd.to_datetime('2020-07-01'),:]
```

```python
df202007.drop(columns=['Date','Website'],inplace=True)
df202007
```

```python
counts = pd.DataFrame({'Counts':df202007.sum()})
counts.sort_values(by=['Counts'],ascending=False,inplace=True)
counts.head(20)
```

```python
counts.to_csv('stock202007.csv')
```
