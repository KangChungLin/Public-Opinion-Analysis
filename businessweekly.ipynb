```python
import requests
from bs4 import BeautifulSoup
import re
import json
import pandas as pd

urls = []


def get_urls():
    for i in range(1,155):
        try:
            res = requests.get(f'https://wealth.businessweekly.com.tw/BChannelSub.aspx?Channel_no=0004&p={i}')
            soup = BeautifulSoup(res.text,'html.parser')
            a_s = soup.select('div.category_art > h2 > a')
            for a in a_s:
                urls.append(a.get('href'))
            
            break
        except:
            print('Error')
            print(f'https://wealth.businessweekly.com.tw/BChannelSub.aspx?Channel_no=0004&p={i}')

get_urls()

url_list = []
title_list = []
author_list = []
date_list = []
content_list = []

def get_content():

    for url in urls:
        try:
            res = requests.get(url+'&p=0')
            soup = BeautifulSoup(res.text,'html.parser')
            url_list.append(url)

            try:
                date_list.append(soup.select_one('div.articlefont > div.article_date > span:nth-child(2)').text)
            except:
                date_list.append('')
            try:
                title_list.append(soup.select_one('div.titxshr > h1').text)
            except:
                title_list.append('')
            try:
                author_list.append(soup.select_one('div.articlefont > div.article_date > span:nth-child(1)').text)
            except:
                author_list.append('')
            try:
                content_list.append(soup.select_one('div.article_main').text)
            except:
                content_list.append('')

            break

        except:
            print(url)

def save_csv():            
    pd_store= pd.DataFrame(columns = ['URL','Date','Title','Author','Content'])
    pd_store['URL'] = url_list
    pd_store['Date'] = date_list
    pd_store['Title'] = title_list
    pd_store['Author'] = author_list
    pd_store['Content'] = content_list

    pd_store.to_csv('test.csv', index=False)
    
# main
get_urls()
get_content()
save_csv()
```
