```python
import pandas as pd
import numpy as np
df = pd.read_csv('final_csv/all_web.csv')
df
```

```python
# 讀取台股名稱列表
stock = pd.read_csv('stock_list/tw_stock_final.csv')
stock
```

```python
def clean_content(df):
    # 'Content'是要清理的欄位名稱(colname) 若欄位名稱不一樣 要修改
    content = df
    
    #下面是清理跟內文無關的文字 像是作者資訊 延伸閱讀等等 
    content = content.replace('-','')
    
    return content
# 一次清理整條column，並回存到原來位置
names = stock['Name'].apply(clean_content)
```

```python
# test
# 先找第一篇文章中的股票名
for name in stock['Name']:
    if name in df['Content'][0]:
        print(name)
```

```python
# 尋找所有文章中的股票名
# 一篇文章同個股票名只算一次
name_list = []
for content in df['Content']:
    stock_set = {name for name in names if name in content}
    name_list.append(' '.join(stock_set))
```

```python
df['Stock'] = name_list
df
```
