```python
import pandas as pd
df = pd.read_csv('businessweekly/weekly_1.csv')
df
```

```python
df['Content'][0]
```

```python
# 只保留中文 數字 英文 標點符號
# 移除特殊符號
df['Content'] = df['Content'].str.replace('[^\u4e00-\u9fa5\u0041-\u005a\u0061-\u007a\u0030-\u0039,。:?!;，！。]+','')
```

```python
df['Content'][0]
```

```python
# 刪除作者資料 與內文無關的資訊
df['Content'] = df['Content'].str.split('作者簡介').str.get(0)
```
