```python
# clean data
# gihun_blog
import pandas as pd
import re

# read file
#df = pd.read_csv('csv/mobile01.csv')

def clean_content(df):
    # 'Content'是要清理的欄位名稱(colname) 若欄位名稱不一樣 要修改
    content = df['Content']
    
    #下面是清理跟內文無關的文字 像是作者資訊 延伸閱讀等等 
    #content = content.split('※')[0].split('好文章')[0].split('《理財財商》')[0].split('趕快登入')[0].split('歡迎加入')[0].replace('圖片來源','').replace('CMoney','')
    content = content.split('大昌證券')[0].split('http')[0].split('康和期貨')[0].split('永豐期貨')[0].split('元富期貨')[0]
    
    content = re.sub(r'\(.{1,15}\)','',content)
    content = re.sub(r'【.{1,10}】','',content)
    #只保留中文字
    content = re.sub(r'[^\u4e00-\u9fa5\u0041-\u005a\u0061-\u007a\u0030-\u0039]','',content)
    
    return content

# 一次清理整條column，並回存到原來位置
df_all['Content'] = df_all.apply(clean_content,axis=1)
```
